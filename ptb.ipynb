```python
import nltk
from nltk.corpus import treebank
from nltk.tree import Tree

# Download necessary resources if not already installed
nltk.download('treebank')

# Get the parsed sentences
parsed_sentences = treebank.parsed_sents()

# Print the first 100 sentences with pretty print
for i, tree in enumerate(parsed_sentences[:100], 1):
    print(f"\nSentence {i}:")
    tree.pretty_print()
```
